```python
import pandas as pd
```

```python
disease = pd.read_csv("https://github.com/deepak-006/diseasePredictor/raw/main/diseasePrediction.csv")
```

```python
disease.head()
```

```python
disease.info(all)
```

```python
disease.columns
```

```python
y = disease['prognosis']
```

```python
X = disease[disease.columns.difference(['prognosis'])]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=25)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
